```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
df.select_dtypes(include="number").head(10)
```

```python
print(f"Jumlah record = {df.shape[0]}")
print(f"Jumlah atribut = {df.shape[1]}")
print(f"Nama atribut = {df.columns.tolist()}")
print(f"Jumlah atribut bertipe angka = {df.select_dtypes(include='number').shape[1]}")
```

```python
for col in df.select_dtypes(include="object"):
    print(f"- {col}: {df[col].nunique()}")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Satisfaction Level", y="Total Spend", data=df, estimator=sum, errorbar=None)
plt.title("Total Spend berdasarkan Satisfaction Level")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi")
plt.show()
```
